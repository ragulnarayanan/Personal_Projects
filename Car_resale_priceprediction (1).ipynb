## Importing the Dataset

```python
import pandas as pd

#reading the data
car = pd.read_csv('/content/car-resale.csv')

#Inspect the data
car.head()
```

## Feature Engineering on the dataset

```python
#Identifying Null values
sumofnull = car.isnull().sum()

#Inspect the data
sumofnull
```

```python
#Identifying dupilcates
duplicated = car.duplicated().sum()

#Inspect the data
duplicated
```

```python
#Dropping duplicates
car.drop_duplicates(inplace = True)

#Inspect the data
duplicated = car.duplicated().sum()
```

```python
#Dropping columns
car.drop(columns = ['Car_Name', 'Engine_Type', 'CC_Displacement','Power(BHP)','Torque(Nm)','Fuel_Tank_Capacity(L)','Transmission','Color','Body_Type'],inplace = True)

car.head()
```

```python
#Cleaning price column
car['Price'] = car['Price'].str.replace(',', '')

#inspect the data
car.head()
```

```python
#Cleaning Mileage_run column
car['Mileage_Run'] = car['Mileage_Run'].str.replace(',','')

#inspect the data
car.head()
```

```python
#Renaming column
car.rename(columns={'Make': 'Company'}, inplace=True)

#inspect the data
car.head()
```

```python
#Encoding categorical features

from sklearn.preprocessing import LabelEncoder

lab_encode = LabelEncoder()
# for columns in ['No_of_Owners','Fuel_Type','Emission','Transmission_Type']:
for columns in ['Transmission_Type']:
    car[columns] = lab_encode.fit_transform(car[columns])


#--- Inspect data ---
car.head()

#Inserted label values for reference
#Noofowners = 0,1,2
#Fuel_type = 0,1,2 {D,P,CNG}
#Transmissiontype = 0,1 {A,M}
#Emmission = 0,1,2 {BS3,BS4,BS5}
```

```python
# Assuming 'car' is your DataFrame

# Convert 'Mileage(kmpl)' column to numeric, coercing errors to NaN
car['Mileage(kmpl)'] = pd.to_numeric(car['Mileage(kmpl)'], errors='coerce')

# Round the values
car['Mileage(kmpl)'] = car['Mileage(kmpl)'].round()

# Convert to float
car['Mileage(kmpl)'] = car['Mileage(kmpl)'].astype(float)

# Display the head of the DataFrame
car.head()
```

```python
# Replace NaN values with a specific value (e.g., 0)
car['Mileage(kmpl)'].fillna(0, inplace=True)

# Convert to integers
car['Mileage(kmpl)'] = car['Mileage(kmpl)'].astype(int)
```

```python
# Assuming 'car' is your DataFrame
object_columns = ['Mileage_Run', 'Mileage(kmpl)', 'Price']

# Convert specific columns to numeric
car[object_columns] = car[object_columns].apply(pd.to_numeric, errors='coerce')

# inspect the DataFrame
dataTypes2 = car.dtypes
dataTypes2
```

## Splitting Data for Regression Modeling

```python
#Building the Machine learning models
from sklearn.model_selection import train_test_split

x = car.drop('Price',axis=1)
Y = car['Price']

x_train, x_test, y_train, y_test = train_test_split(x, Y, test_size=0.2, random_state=4)
```

## One-Hot Encoding Categorical Features

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer

# Instantiate OneHotEncoder
ohe = OneHotEncoder()

# Fit OneHotEncoder on the categorical columns in 'x'
ohe.fit(x[['Model', 'Company', 'Emission', 'Fuel_Type', 'No_of_Owners']])

# Create a column transformer using make_column_transformer
column_trans = make_column_transformer(
    (OneHotEncoder(categories=ohe.categories_), ['Model', 'Company', 'Emission', 'Fuel_Type', 'No_of_Owners']),
    remainder='passthrough'
)
```

## Building and Evaluating Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score

# Instantiate LinearRegression model
lr = LinearRegression()

# Create a pipeline with column transformer and LinearRegression model
pipe = make_pipeline(column_trans, lr)

# Fit the pipeline on the training data
pipe.fit(x_train, y_train)

# Generate predictions on the test data using the fitted pipeline
y_pred = pipe.predict(x_test)

# Calculate the R-squared score by comparing predicted values with actual values
lr_score = r2_score(y_test, y_pred)
lr_score
```

Building and Evaluating Lasso Regression Model

```python
from sklearn.linear_model import Lasso

# Instantiate Lasso model
model_lasso = Lasso(alpha=1,max_iter=10000,tol=0.001)

# Create a pipeline with column transformer and Lasso model
pipe2 = make_pipeline(column_trans, model_lasso)

# Fit the pipeline on the training data
pipe2.fit(x_train, y_train)

# Generate predictions on the test data using the fitted pipeline
y_pred = pipe2.predict(x_test)

# Calculate the R-squared score by comparing predicted values with actual values
lasso_score = r2_score(y_test, y_pred)
lasso_score
```

## Building and Evaluating Random Forest Regression Model

```python
from sklearn.ensemble import RandomForestRegressor

# Instantiate RandomForestRegressor
rf = RandomForestRegressor(n_estimators=10)

# Create pipeline with column transformer and RandomForestRegressor
pipe1 = make_pipeline(column_trans, rf)

# Fit the pipeline on the training data
pipe1.fit(x_train, y_train)

# Generate predictions on the test data using the fitted pipeline
y_pred = pipe1.predict(x_test)

# Calculate the R-squared score by comparing predicted values with actual values
rf_score = r2_score(y_test, y_pred)
rf_score
```

## Building and Evaluating Decision Tree Regression Model

```python
from sklearn.tree import DecisionTreeRegressor

#instantiate Decision Tree Regression
dtr = DecisionTreeRegressor()

pipe3 = make_pipeline(column_trans, dtr)

pipe3.fit(x_train, y_train)

y_pred = pipe3.predict(x_test)

rf_score = r2_score(y_test, y_pred)
rf_score
```
